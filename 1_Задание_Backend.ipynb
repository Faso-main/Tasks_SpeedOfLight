# Реализация холодильника по принципам SOLID

## Задача
- Разработать программную модель холодильника на выбранном вами языке, реализующую принципы ООП и SOLID.
- Самого настоящего холодильника, который можно открывать и закрывать, убирать в него что-то или доставать, простор фантазии бесконечен.

## Критерии оценки
1. **Полное соблюдение принципов SOLID:**
   - SRP: Единственная ответственность
   - OCP: Открытость/закрытость
   - LSP: Подстановка Лисков
   - ISP: Разделение интерфейсов
   - DIP: Инверсия зависимостей

2. **Качество архитектуры**
