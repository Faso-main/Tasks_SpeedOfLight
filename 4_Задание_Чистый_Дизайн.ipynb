# Постер «Набор в команду»

## Цель задания
Проверить ваше умение работать с композицией, типографикой и цветом в Figma, а также вашу способность следовать инструкции и креативно подходить к решению дизайнерской задачи.

## Суть задачи
Вам необходимо создать творческий и современный постер, который анонсирует набор в новую команду дизайнеров/разработчиков название команды «SpeedOfLight».

## Технические требования (Обязательные к выполнению)

**Инструмент** - Figma. (https://www.figma.com/downloads/)

**Размер артборда** - A3 (297 x 420 мм) или 1240 x 1754 px. (Стандарт постеров https://slovodelo.ru/blog/hints/razmery-plakatov-dlya-pechati-chto-nuzhno-znat-o-standartnykh-parametrakh/)

**Содержание текста**
- Заголовок - Ищем таланты (пример) (На своё усмотрение)
- Подзаголовок - Стань частью команды [Название команды] (пример) (На своё усмотрение)
- Основной текст - Мы — молодая команда СамГТУ, занимающаяся созданием цифровых продуктов. Вместе мы решаем сложные задачи реальных и партнёрских кейсов. Готов стать частью команды? (пример) (На своё усмотрение)

**Шрифты** - Используйте не более двух шрифтовых семейств. Рекомендую выбрать из Google Fonts

**Цветовая палитра** - Используйте не более 3-4 цветов, включая основной, акцентный и фоновый. (https://ki-paris.ru/blog/palitra-sochetanie-tsvetov/?utm_referrer=https%3A%2F%2Fwww.google.com%2F) + (https://hudozhnik.online/blog/chvetovoi_krug)

**Элементы** - Для украшения можно использовать простые геометрические фигуры (круги, прямоугольники, линии), иконки или текстуры. Не используйте стоковые фотографии (из инета не круто)

**Сетка** - Плакат должен быть выверен по композиции. Рекомендую использовать сетку для выравнивания элементов. (https://www.figma.com/community/plugin/1525764954610971476/figma-figma-inspire) + (ПРИМЕР:https://www.figma.com/community/plugin/1525764954610971476/figma-figma-inspire)

## Критерии оценки

**Композиция (40%)**
- Логичное и четкое визуальное разделение заголовка, подзаголовка, основного текста и призыва к действию
- Сбалансированное расположение элементов, использование воздуха (пространства)
- Наличие четкой точки фокуса, которая сразу привлекает внимание

**Работа со шрифтами (30%)**
- Удачное сочетание шрифтов (например, шрифт с засечками для заголовка + гротеск для текста, или наоборот)
- Создание контраста за счет размера, начертания (жирный, обычный) и интерлиньяжа (межстрочного расстояния)

**Цвет и контраст (20%)**
- Гармонично подобранная цветовая палитра
- Достаточный контраст между текстом и фоном для удобочитаемости
- Акцентный цвет используется осмысленно для выделения ключевой информации

**Следование ТЗ и аккуратность (10%)**
- Выполнены все технические требования
- Файл в Figma аккуратно организован - слои и фреймы имеют понятные названия, элементы сгруппированы
