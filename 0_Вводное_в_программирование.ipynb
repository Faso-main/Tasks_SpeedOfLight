## Критерии оценки
1. **Полное соблюдение принципов SOLID:**
   - SRP: Единственная ответственность
   - OCP: Открытость/закрытость
   - LSP: Подстановка Лисков
   - ISP: Разделение интерфейсов
   - DIP: Инверсия зависимостей
